#### From the Fitbit data, we want to visually understand:

```python
# MEDIAN STEPS BY MONTH_ENCODED
median_steps = fitbit.groupby("Months_encoded")[["Steps"]].median()
fig, ax = plt.subplots(figsize=(10,5))
print(ax.plot(median_steps))
# your code here

# MEDIAN STEPS BY WORK_OR_WEEKEND
median_w_or_w = fitbit.groupby("Work_or_Weekend")["Steps"].median()
fig, ax = plt.subplots(figsize=(10,5))
median_w_or_w.plot.bar()
# your code here
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
columns = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']

f, ax = plt.subplots(3,1, figsize= (8,18))
values = fitbit['Steps'].values

index = 0
for c in columns:
    names = fitbit[c].values
    plt.sca(ax[index])
    plt.scatter(names,values)
    plt.title(c)
    index += 1
```
